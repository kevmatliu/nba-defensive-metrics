```python
import pandas as pd
from nba_api.stats.endpoints import draftcombineplayeranthro
```

```python
draft_combine_ = []

for year in range(2003, 2026):
    draft_combine = draftcombineplayeranthro.DraftCombinePlayerAnthro(season_year=year)
    draft_combine_.append(draft_combine.get_data_frames()[0])

draft_combine_df = pd.concat(draft_combine_)
draft_combine_df
```

```python
draft_combine_df.to_csv("data/wingspan.csv")
```
