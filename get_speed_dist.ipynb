```python
import pandas as pd
from nba_api.stats.endpoints import leaguedashptstats
```

```python
league_dash = leaguedashptstats.LeagueDashPtStats(player_or_team="Player")
league_dash_df = league_dash.get_data_frames()[0]

league_dash_df
```

```python
league_dash_df["AVG_DIST_MILES_DEF"] = league_dash_df["DIST_MILES_DEF"] / league_dash_df["GP"]
league_dash_df["AVG_DIST_MILES_OFF"] = league_dash_df["DIST_MILES_OFF"] / league_dash_df["GP"]
league_dash_df["AVG_DIST_DEF_OFF_RATIO"] = league_dash_df["AVG_DIST_MILES_DEF"] / league_dash_df["AVG_DIST_MILES_OFF"]

league_dash_df["AVG_DIST_MILES_DEF_MIN"] = league_dash_df["DIST_MILES_DEF"] / league_dash_df["MIN"]
league_dash_df["AVG_DIST_MILES_OFF_MIN"] = league_dash_df["DIST_MILES_OFF"] / league_dash_df["MIN"]

league_dash_df["AVG_MIN"] = league_dash_df["MIN"] / league_dash_df["GP"]

league_dash_df
```

```python
league_dash_df.to_csv("data/speed-dist.csv")
```
