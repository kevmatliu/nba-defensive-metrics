```python
import pandas as pd
```

```python
defensive_impact_df = pd.read_csv('data/defensive-impact.csv')
defensive_rebounding_df = pd.read_csv('data/defensive-rebounding.csv')
speed_dist_df = pd.read_csv('data/speed-dist.csv')
wingspan_df = pd.read_csv('data/wingspan.csv')
```

```python
defensive_impact_df.rename(columns={'Player': 'PLAYER_NAME', 'Team': 'TEAM_ABBREVIATION'}, inplace=True)
defensive_rebounding_df.rename(columns={'PLAYER': 'PLAYER_NAME', 'TEAM': 'TEAM_ABBREVIATION'}, inplace=True)
```

```python
speed_dist_df = speed_dist_df[speed_dist_df['GP'] >= 15]
speed_dist_df = speed_dist_df[speed_dist_df['MIN'] >= 15]
len(speed_dist_df)
```

```python
'''
AVG_DIST_DEF_OFF_RATIO (speed-dist.csv, AVG_DIST_DEF_OFF_RATIO)
AVG_DIST_MILES_DEF_MIN (speed-dist.csv, AVG_DIST_MILES_DEF_MIN)
DEF_RIM_FGA (defensive-impact.csv, DFGA)
DREB_CONTEST_PCT (defensive-rebounding.csv, ContestedDREB%)
DREB_CHANCES (defensive-rebounding.csv, DREBChances)
AVG_DREB_DISTANCE (defensive-rebounding.csv, AVGDREBDistance)
WINGSPAN (wingspan.csv, WINGSPAN)
'''


master_df = speed_dist_df[['PLAYER_NAME', 'TEAM_ABBREVIATION']].copy()
master_df = master_df.merge(speed_dist_df[['PLAYER_NAME', 'PLAYER_ID', 'AVG_DIST_DEF_OFF_RATIO', 'AVG_DIST_MILES_DEF_MIN']], on='PLAYER_NAME', how='left')
master_df = master_df.merge(defensive_rebounding_df[['PLAYER_NAME', 'ContestedDREB%', 'DREBChances', 'AVGDREBDistance']], on='PLAYER_NAME', how='left')
master_df = master_df.merge(defensive_impact_df[['PLAYER_NAME', 'DFGA']], on='PLAYER_NAME', how='left')

# master_df = master_df.merge(wingspan_df[['PLAYER_ID', 'WINGSPAN']], on='PLAYER_ID', how='left')

master_df
```

```python
master_df = master_df.dropna()

master_df['AVG_DIST_DEF_OFF_RATIO'] = pd.to_numeric(master_df['AVG_DIST_DEF_OFF_RATIO'])
master_df['AVG_DIST_MILES_DEF_MIN'] = pd.to_numeric(master_df['AVG_DIST_MILES_DEF_MIN'])
master_df['DFGA'] = pd.to_numeric(master_df['DFGA'])
master_df['ContestedDREB%'] = pd.to_numeric(master_df['ContestedDREB%'])
master_df['DREBChances'] = pd.to_numeric(master_df['DREBChances'])
master_df['AVGDREBDistance'] = pd.to_numeric(master_df['AVGDREBDistance'])

master_df.to_csv('data/master.csv', index=False)
```
